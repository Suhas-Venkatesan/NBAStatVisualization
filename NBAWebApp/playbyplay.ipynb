```python
from nba_api.stats.endpoints import playbyplayv2
import plotly.graph_objects as go
import pandas as pd
from nba_api.stats.endpoints import leaguegamefinder
import datetime
pbp = playbyplayv2.PlayByPlayV2("0041700401")
pbp = pbp.get_data_frames()[0]
pbp = pbp.loc[pbp['EVENTMSGTYPE'] == 1]

#if the period is 4 or less
pbp['secs'] = pbp['PCTIMESTRING'].str.split(":").str[0].astype(int).rsub(12).mul(60).sub(pbp['PCTIMESTRING']
                .str.split(":").str[1].astype(int)).add(pbp['PERIOD'].sub(1).mul(720))
#if the period is more than 4

pbp.loc[(pbp['PERIOD'] > 4), 'secs'] = pbp['PCTIMESTRING'].str.split(":").str[0].astype(int).rsub(5).mul(60).sub(pbp['PCTIMESTRING'].str.split(":").str[1].astype(int)).add(2880).add(pbp['PERIOD'].sub(5).mul(300))


##
XAXIS = pbp['secs'].floordiv(60).astype(str)
XAXIS = XAXIS.add(":")
XAXIS = XAXIS.add(pbp['secs'].mod(60).astype(str))
XAXIS = pd.to_datetime(XAXIS, format = "%M:%S").dt.time
pbp['XAXIS'] = XAXIS


fig = go.Figure()
values = list(range(-40,40))

fig.add_trace(go.Scatter(
    x=pbp['XAXIS'],
    y= pbp['SCOREMARGIN'],
    name = '<b>No</b> Gaps', # Style name/legend entry with html tags
    connectgaps=True,
    mode='lines+markers',
    marker=dict(
        size=13,
        cmax=40,
        cmin=-40,
        
    )
))

fig.update_layout({
'plot_bgcolor': 'rgba(0, 0, 0, 0)',
'paper_bgcolor': 'rgba(0, 0, 0, 0)',
})



fig.update_layout(
    title={
        'text': "Game",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'},
    xaxis_title="Time in Minutes  ",
    yaxis_title="Score Margin (home-away)",
    font=dict(
        family="Verdana",
        size=18,
        color="#7f7f7f"
    ),
    
)

fig.add_shape(
        # Line Horizontal
            type="line",
            x0=-1,
            y0=0,
            x1=68,
            y1=0,
            line=dict(
                color="#000000",
                width=2,
                dash="dot",
            ),
    )

fig.update_layout(
    hoverlabel=dict(
        bgcolor="white", 
        font_size=16, 
        font_family="Rockwell"
    ),
    annotations=[
            dict(
                x=datetime.datetime(year=1900, month=1, day=1, minute=59, second=0),
                y=15,
                xref="x",
                yref="y",
                text="HOME",
            )
        ]
)
```
